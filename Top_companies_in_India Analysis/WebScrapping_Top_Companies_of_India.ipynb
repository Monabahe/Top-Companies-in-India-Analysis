# Scraping details of top companies of India by Revenue
#### LINK : https://en.wikipedia.org/wiki/List_of_largest_companies_in_India
**Project Objective:**

The objective of this project is to scrape and analyze the details of the top revenue-generating companies in India from the Wikipedia page "List of largest companies in India by revenue." The data collected will include the company names, industries, revenue figures, and other relevant information. This dataset will be used for data analysis to identify trends, patterns, and insights into the economic landscape of India.

**Web Scraping Method:**

**Tools and Libraries:**

**Python**: The main programming language for this project.

**BeautifulSoup:** A library for parsing HTML and XML documents.

**Pandas:** A data manipulation and analysis library.

 **Importing the libraries**

```python
from bs4 import BeautifulSoup
import requests
# for creating a dataframe
import pandas as pd
```

**Getting the response**

```python
#Driver assigning and getting url requests

url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_India"

header = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36"}
response = requests.get(url, headers = header)

print(response)
```

```python
soup = BeautifulSoup(response.text, "html")
```

## Table 1

**Extracting the table 1 of year 2023 companies by their revenue in US$**

```python
Table_1 = soup.find_all("table")[0] #index 0 for first table
# for i in Table_1:
#     print(i.text.strip())
```

```python
#scraping the columns heading from the table_1

columns_headings = Table_1.find_all("th")

column = []

for headings in columns_headings:
    column.append(headings.text.strip())
column
```

**Importing the columns names inside the dataframe by using pandas**

```python
Table_2023 = pd.DataFrame(columns = column)
Table_2023
```

```python
Whole_table_1 = Table_1.find_all("tr")[1:]

for each_row in Whole_table_1:
    
    td = each_row.find_all("td")
    all_data = []
    
    for row in td:
        
        all_data.append(row.text.strip())
        
    len_ = len(Table_2023)
    Table_2023.loc[len_] = all_data
```

```python
Table_2023
```
